```python
import html

sentences = []
should_print = False
with open("The_New_York_Times2015-05-01_15-02.txt") as f:
    for line in f.readlines():
        if line.startswith("HIGHLIGHT") or line.startswith("LENGTH"):
            should_print = True
            sentence = []
            continue
        elif line.startswith("LOAD-DATE") or line.startswith("URL"):
            should_print = False
            sentences += sentence
            continue
        if should_print and not line.startswith("HIGHLIGHT"):
            line_strip = line.strip()
            if (line_strip
                and line_strip[0] not in "0123456789"):
                sentence.append(html.unescape(line_strip))
```

```python
import random
import nltk

text = ". ".join(sentences)
tokenized_sentences = nltk.sent_tokenize(text)
clean_sentences = []
for sent in tokenized_sentences:
    len_sent = len(sent)
    if (len_sent > 50
        and "..." not in sent
        and not sent.startswith("(")
        and "@" not in sent
        and "http" not in sent
        and ".org" not in sent
        and ".com" not in sent
        and not sent[0].islower()):
        sent = sent.replace("''", '"')
        sent = sent.replace("..", ".")
        clean_sentences.append(sent[:-1])

random.shuffle(clean_sentences)
clean_sentences
```

```python
with open("control_sentences.txt", "w") as outfile:
    outfile.write("\n\n".join(clean_sentences))
```
